# Matplotlib

Matplotlib allows us to create plots and graphs. This enables us to visualize data that we have stored as a dataframe. Let us read a csv file as a data frame and produce some plots:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://drive.google.com/uc?export=download&id=19vb0TTdUEXm8cdY8zkH8ZNfQms4czzvW')
df.head(10)
```

We have a data frame with three columns, one of which is year. Let us produce a plot of total deaths each year:

```python
plt.plot(df['Year'], df['Total'])
```

Let us add a title to our graph:

```python
plt.plot(df['Year'], df['Total'])
plt.title("Total deaths each year")
```

Let us make the title bigger and in red:

```python
plt.plot(df['Year'], df['Total'])
plt.title("Total deaths each year", fontsize=24, color='r')
```

Let us also add titles to the x and y axes:

```python
plt.plot(df['Year'], df['Total'])
plt.title("Total deaths each year", fontsize=24, color='r')
plt.xlabel("Year")
plt.ylabel("Deaths")
```

We can change the color of the line from the default blue to red and make the line dashed, and make the connected data points triangles:

```python
df.columns
```

```python
plt.plot(df['Year'], df['Total'], color='black', marker='x', linestyle='-.')
plt.title("Total deaths each year", fontsize=24, color='r')
plt.xlabel("Year")
plt.ylabel("Deaths")
```

Let us now add a plot of alcogol-related deaths to the plot. This data is in the column `Alcohol-related`:

```python
plt.plot(df['Year'], df['Total'], color='red', marker='^', linestyle='--')
plt.plot(df['Year'], df['Alcohol-related'], color='g', marker='o', linestyle='-.')
plt.title("Total deaths each year", fontsize=24, color='r')
plt.xlabel("Year")
plt.ylabel("Deaths")
```

Because we have two lines, it would be a good idea to use a legend to distinguish between them. We can do this by adding a legend to the plot using the `plt.legend()` function. However, we need to specify the labels for the lines. We can do this by adding a label to each line using the `label` argument in the `plt.plot()` function:

```python
plt.plot(df['Year'], df['Total'], color='red', marker='^', linestyle='--', label='Total deaths')
plt.plot(df['Year'], df['Alcohol-related'], color='g', marker='o', linestyle='-.', label='Alcohol-related deaths')
plt.title("Total deaths each year", fontsize=24, color='r')
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.legend()
```

Instead of specifying the color, marker, and line style in separate arguments, you can specify them all in a single string argument:

```python
plt.plot(df['Year'], df['Total'], 'r^--', label='Total deaths', )
plt.plot(df['Year'], df['Alcohol-related'], 'go-.', label='Alcohol-related deaths')
plt.title("Total deaths each year", fontsize=24, color='r')
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.legend()
```

We can also add gridlines to the plot:

```python
plt.grid(linestyle='--', color='green')
plt.plot(df['Year'], df['Total'], 'r^--', label='Total deaths', )
plt.plot(df['Year'], df['Alcohol-related'], 'go-.', label='Alcohol-related deaths')
plt.title("Total deaths each year", fontsize=24, color='r')
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.legend()
```

We can save our plot as an image file:

```python
plt.savefig('my_plot.png')
```

## A scatter plot example

Let us now see how we can produce a scatter plot where the points are colored according to their class:

```python
# create a dataframe that contains only the data from before 2000
df_old = df[df['Year'] < 2000]
# create a dataframe that contains only the data from 2000 and after
df_new = df[df['Year'] >= 2000]
# plot the old data
plt.scatter(df_old['Total'], df_old['Alcohol-related'], label='Old data')
# plot the new data
plt.scatter(df_new['Total'], df_new['Alcohol-related'], label='New data')
plt.legend()
```

# Multiple plots

In addition to drawing two graphs on the same plot, we can also produce two separate plots. To do that, we use the `subplot` function. This function takes three arguments: the number of rows, the number of columns, and the index of the plot we want to draw.

```python
# create a dataframe that contains only the data from before 2000
df_old = df[df['Year'] < 2000]
# create a dataframe that contains only the data from 2000 and after
df_new = df[df['Year'] >= 2000]
# plot the old data
plt.subplot(1, 2, 1)
plt.scatter(df_old['Total'], df_old['Alcohol-related'], label='Old data')
plt.title("Old data")
# plot the new data
plt.subplot(1, 2, 2)
plt.scatter(df_new['Total'], df_new['Alcohol-related'], label='New data')
plt.title("New data")
```
